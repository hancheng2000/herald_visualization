```python
import csv

def convert_txt_to_csv(input_file, output_file):
    with open(input_file, 'r') as txt_file:
        lines = txt_file.readlines()
        
    # Assuming the first line is the header
    data_lines = lines[1:]
    
    with open(output_file, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        
        for line in data_lines:
            columns = line.strip().split()  # Adjust this if the delimiter is different (e.g., tab or comma)
            if len(columns) >= 3:
                csv_writer.writerow([columns[1], columns[2]])

# import pandas as pd

# # Input and output file paths
# input_file = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_1to1_KS6Lgraphite/RR_ASC_0mins_CoF3/Differences_240827-CoF3_1to1_KS6Lgraphite-0min_milled-20-70deg_40min.ASC'
# output_file = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_1to1_KS6Lgraphite/RR_ASC_0mins_CoF3/Differences_240827-CoF3_1to1_KS6Lgraphite-0min_milled-20-70deg_40min.csv'
# # Read ASC file (adjust delimiter if needed)
# df = pd.read_csv(input_file, delimiter='\s+')  # Try '\s+' for space-separated

# # Save as CSV with commas as the separator
# df.to_csv(output_file, index=False, sep=',')
```

```python

from matplotlib.lines import Line2D

# List of CSV files
    
zero_min = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/240827-CoF3_1to1_KS6Lgraphite-0min_milled-20-70deg_40min.csv' 
hundred_min = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/240827-CoF3_1to1_KS6Lgraphite-100min_milled-20-70deg_40min.csv'
CoF3_ref = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_reference.csv'
CoF2_ref =             '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF2_reference.csv'
graphite_ref =   '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/Graphite_reference_final.csv'
fit_100 = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_100min_fitted_data.csv'
fit_0 =  '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_1to1_KS6Lgraphite/RR_ASC_0mins_CoF3/Calculated_Profile_240827-CoF3_1to1_KS6Lgraphite-0min_milled-20-70deg_40min.csv'
differences_100 = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/differences_100mins.csv'
differences_0 = '/Users/deliaharms/git/herald_visualization/herald_visualization/Delia_Scripts/CoF3/CoF3_1to1_KS6Lgraphite/RR_ASC_0mins_CoF3/Differences_240827-CoF3_1to1_KS6Lgraphite-0min_milled-20-70deg_40min.csv'  # Update with actual file names

csv_files = [zero_min,hundred_min, CoF3_ref,CoF2_ref, graphite_ref, fit_100, fit_0, differences_100, differences_0]

colors = ["#1f77b4", "#ff7f0e", "#2ca02c", '#9467bd','black','black','black','black','black']  # Colors for each CSV
legend_labels = ['Milled 0 Minutes', 'Milled 100 Minutes', 'CoF3 Reference','CoF2 Reference','Graphite Reference',None,None,None,None]

plt.figure(figsize=(10, 5))

# Loop through each CSV and plot
for i, csv_file in enumerate(csv_files):
    
    data = pd.read_csv(csv_file)
    print(f'Data from {csv_file}:')
    print(data.head())
    
    x_column = '2Theta'  # Change to your x-axis column name
    y_column = 'Intensity'  # Change to your y-axis column name
    
    if csv_file in {CoF2_ref, CoF3_ref, graphite_ref}:
        plt.bar(data['2Theta'], 4.5*data['Intensity'], width = 0.25,color=colors[i % len(colors)], label = legend_labels[i])
    elif csv_file in {zero_min}:
        plt.plot(data['2Theta'], 5000+data['Intensity'], marker='o', markersize=.01, linestyle='-', color=colors[i % len(colors)], label = legend_labels[i])
    elif csv_file in {fit_0}:
        plt.plot(data['2Theta'], 5000+data['Intensity'], marker='o', markersize=.01, linestyle='-', color=colors[i % len(colors)], label = legend_labels[i])
    elif csv_file in {differences_0}:
        plt.plot(data['2Theta'], 10000+data['Intensity'], marker='o', markersize=.01, linestyle='-', color=colors[i % len(colors)])
    elif csv_file in {differences_100}:
        plt.plot(data['2Theta'], 5000+data['Intensity'], marker='o', markersize=.01, linestyle='-', color=colors[i % len(colors)])
    else:
        plt.plot(data['2Theta'], data['Intensity'], marker='o', markersize=.01, linestyle='-', color=colors[i % len(colors)], label = legend_labels[i])

custom_handles = [Line2D([0], [0], color=color, lw=3) for color in colors]
legend_labels_for_plot = [label for label in legend_labels if label is not None]

plt.xlabel('2 Theta (°)')
plt.ylabel('Intensity (cts)')
plt.title('XRD of CoF3 and Graphite 1:1 Molar')
plt.xlim(20,70)
plt.legend(handles=custom_handles, labels=legend_labels_for_plot)
plt.grid()
plt.show()
```
